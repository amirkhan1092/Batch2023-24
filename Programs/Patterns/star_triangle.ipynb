```python
# triangle using 
rows = int(input('Enter the number of rows  '))
for i in range(1, rows+1):
    print('*')
```

```python
# rows = int(input('Enter the rows '))
rows = 5
for i in range(1, rows+1):
    for j in range(rows-i+1):
        print('&', end='')
    for j in range(i-1):
        print('#', end='')
    print()
```
