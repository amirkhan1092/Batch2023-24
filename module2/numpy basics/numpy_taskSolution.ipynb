```python
# Task 4

n, m = list(map(eval, input().split()))
M1 = []
for i in range(n):
    M1.append(list(map(eval, input().split())))

M2 = []
for i in range(n):
    M2.append(list(map(eval, input().split())))

A = np.array(M1)
B = np.array(M2)

print(np.add(A,B))
print(np.subtract(A,B))
print(np.multiply(A,B))
print(np.floor_divide(A,B))
print(np.mod(A,B))
print(np.power(A,B))
```
